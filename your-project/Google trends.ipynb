```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dates = pd.read_csv("multiTimeline.csv")
```

```python
dates["Semana"] = pd.to_datetime(dates['Semana'],infer_datetime_format=True)
```

```python
dates.dtypes
```

```python
sns.set_style("white")

sns.set_context("paper")
sns.lineplot(x="Semana", y="fake news: (A nível mundial)", data=dates, color = "indianred", markers=True)
plt.xticks(rotation=15)
plt.xlabel('Time(years)', fontsize=12)
plt.ylabel('Trend(%)', fontsize=12)
plt.title('Relation between google searches and time', fontsize=16)
plt.show()

```
